```python
def issue_books(student_name, books):
  print(f"\n{student_name} borrowed:")
  for book in books:
    print("-", book)

def calculate_fine(days_late):
  fine_per_day = 5 # Assuming fine is 5 per day
  return days_late * fine_per_day

def student_report(student_name, books, days_late):

  issue_books(student_name, books)
  fine = calculate_fine(days_late)
  print(f"Fine to be paid: ${fine}\n")
```
